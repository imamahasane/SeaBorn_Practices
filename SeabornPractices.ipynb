### Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.
Seaborn is a library for making statistical graphics in Python. It builds on top of matplotlib and integrates closely with pandas data structures.

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
fmri = sns.load_dataset('fmri')
fmri.head()
```

# SeaBorn Line Plot
1. lineplot(p1, p2, p3)
2. p1 = X-label, p2 = Y-label, p3 = data

```python
sns.lineplot(x = 'timepoint', y = 'signal', data = fmri)
plt.show()
```

### Grouping data with 'hue'

```python
sns.lineplot(x = 'timepoint', y = 'signal', data = fmri, hue = 'event')
plt.show()
```

### Adding Styles

```python
sns.lineplot(x = 'timepoint', y = 'signal', data = fmri, hue = 'event', style = 'event')
plt.show()
```

### Adding Markers

```python
sns.lineplot(x = 'timepoint', y = 'signal', 
             hue = 'event', style = 'event',
            markers = True, data = fmri)
plt.show()
```

# SeaBorn Histogram/ Distplot

```python
d = pd.read_csv("diamonds.csv")
d
```

```python
d.head()
```

```python
sns.distplot(d['price'])
plt.show()
```

### hist

```python
sns.distplot(d['price'], hist = False)
plt.show()
```

### bins

```python
sns.distplot(d['price'], color = 'green', bins = 10)
plt.show()
```

### kde

```python
sns.distplot(d['price'], color = 'green', bins = 10, kde = False)
plt.show()
```

### vertical

```python
sns.distplot(d['price'], color = 'green', vertical = True)
plt.show()
```

```python
sns.distplot(d['price'], color = 'green', bins = 10, kde = False, vertical = True)
plt.show()
```
